```python
import os 
import numpy as np
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as image
import glob
```

```python
images_fp = './images'
```

```python
image_names =[os.path.basename(file) for file in glob.glob(os.path.join(images_fp,'*.jpg'))]
```

```python
image_names
```

```python
len(image_names)
```

```python
labels = [' '.join(name.split('_')[:-1]) for name in image_names]
```

```python
labels
```

```python
def label_encode(label):
    if label == 'abyssinian': return 0
    elif label == 'american shorthair' : return 1
    elif label == 'beagle' : return 2
    elif label =='boxer' : return 3
    elif label == 'chihuahua' : return 4
    elif label == 'corgi' : return 5
    elif label == 'dachshund' : return 6
    elif label == 'german shepherd' : return 7
    elif label == 'golden retriever' : return 8
    elif label == 'husky' : return 9
    elif label == 'labrodar' : return 10
    elif label == 'maine coon' : return 11
    elif label =='mumbai cat' : return 12
    elif label == 'persian cat' : return 13
    elif label == 'pomeranian' : return 14
    elif label =='pug' : return 15
    elif label == 'ragdoll cat' : return 16
    elif label == 'rottwiler' : return 17
    elif label == 'shiba inu' : return 18
    elif label == 'siamese cat' : return 19
    elif label == 'sphynx' : return 20
    elif label =='yorkshire' : return 21
    elif label == 'bulldog' : return 22
```

```python
from tensorflow.keras.preprocessing.image import load_img, img_to_array
```

```python
features = []
labels = []
IMAGE_SIZE = (224, 224)

for name in image_names:
    label =' '.join(name.split('_')[:-1])
    label_encoded = label_encode(label)
    if label_encoded != None:
        img = load_img(os.path.join(images_fp, name))
        img = tf.image.resize_with_pad(img_to_array(img,dtype='uint8'),*IMAGE_SIZE).numpy().astype('uint8')
        image = np.array(img)
        features.append(image)
        labels.append(label_encoded)
```

```python
features
```

```python
labels
```

```python
features_array = np.array(features)
labels_array = np.array(labels)
```

```python
labels_one_hot = pd.get_dummies(labels_array)
```

```python
labels_one_hot
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train, y_test = train_test_split(features_array, labels_one_hot, test_size = 0.2, random_state = 42)
```

```python
x_train , x_val , y_train , y_val = train_test_split(x_train, y_train, test_size = 0.25 , random_state = 1)
```

```python
from tensorflow.keras import layers , Input , Model
from tensorflow.keras.models import Sequential 
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.applications.resnet50 import preprocess_input as pp_I
from tensorflow.keras.layers import RandomFlip, RandomRotation, Dense, Dropout
from tensorflow.keras.losses import CategoricalCrossentropy
from tensorflow.keras.optimizers import Adam
```

```python
data_augmentation = Sequential([RandomFlip("horizontal_and_vertical"), RandomRotation(0.2)])
prediction_layers = Dense(21 , activation = 'softmax')
```

```python
resnet_model = ResNet50(include_top=False, pooling='avg',weights='imagenet')
resnet_model.trainable = False
preprocess_input = pp_I
```

```python
inputs = Input(shape=(224,224,3))
x = data_augmentation(inputs)
x = preprocess_input(x)
x = resnet_model(x , training = False)
x = Dropout(0.2)(x)
outputs = prediction_layers(x)
model = Model(inputs, outputs)
```

```python
model.compile(optimizer=Adam(), loss=CategoricalCrossentropy(), metrics=['accuracy'])
```

```python
model_history = model.fit(x =x_train, y = y_train, validation_data=(x_val,y_val), epochs=10)
```

```python
acc = model_history.history['accuracy']
val_acc = model_history.history['val_accuracy']
loss = model_history.history['loss']
val_loss = model_history.history['val_loss']
```

```python
epochs_range = range(10)
plt.figure(figsize=(15,8))

plt.subplot(1,2,1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc , label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1,2,2)
plt.plot(epochs_range, loss , label = 'Training Loss')
plt.plot(epochs_range, val_loss , label= 'Validation Loss')
plt.legend(loc = 'upper right')
plt.title('Training and Validation loss')
```

```python
model.evaluate(x_test , y_test)
```
